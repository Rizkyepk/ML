# Data cleaning : Drop Column

```python
to_drop =[
    'Location ISO Code',
    'City or Regency',
    'Time Zone',
    'Total Urban Villages',
    'Total Rural Villages',
    'Area (km2)',
    'Longitude', 'Latitude',
    'New Cases per Million',
    'Total Cases per Million',
    'New Deaths per Million',
    'Total Deaths per Million',
    'Total Deaths per 100rb',
    'Location Level',
    'Special Status',

]

df.drop(columns=to_drop, inplace=True)

df.info()
```

handling missing data

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
# List kolom yang memiliki nilai null
kolom_null = ['Province', 'Island', 'Total Cities', 'Growth Factor of New Cases', 'Growth Factor of New Deaths']

# Menghapus baris dengan nilai null
df = df.dropna(subset=kolom_null)
```

```python
df.info()
```

Merapikan Fields
dalam Data

```python
# Memisahkan tanggal menjadi tanggal, bulan, dan tahun, kemudian mengambil tahunnya saja
df['Date'] = df['Date'].apply(lambda x: int(str(x).split('/')[2]) if '/' in str(x) else x)

# Menampilkan 5 baris pertama data setelah perubahan
df.head()
```

```python
# Mengubah kolom "Date" menjadi nilai timestamp (float)
df['Date'] = df['Date'].astype(int)
```

```python
df['Date'].is_unique
```

```python
df = df.set_index('Date')
df.head()
```
